```python
# Import data sets
md1 = pd.read_csv('data/2/MissingData1.txt', delimiter='\s+', header=None)
md2 = pd.read_csv('data/2/MissingData2.txt', delimiter='\s+', header=None)
md3 = pd.read_csv('data/2/MissingData3.txt', delimiter='\s+', header=None)
```

```python
n = md1[2][3]
n
```

```python
# Find all values in md1, md2, md3 that = 1.00000000000000e+99 & replace with NaN
for i in md1.keys():
    md1[i] = md1[i].replace(n, np.nan)
    
for i in md2.keys():
    md2[i] = md2[i].replace(n, np.nan)

for i in md3.keys():
    md3[i] = md3[i].replace(n, np.nan)    
```

```python
# Use fast_knn to fill NaN values in md1, md2, md3
md1 = fast_knn(md1.values, k=50)
```
